Yearly data

```python
sales_data = pd.DataFrame(sales)
# Convert date columns to datetime
sales_data['Invoice Date'] = pd.to_datetime(sales_data['Invoice Date'])
sales_data['DateKey'] = pd.to_datetime(sales_data['DateKey'], format='%m/%d/%Y')

# Extract year, month, and quarter from Invoice Date
sales_data['Invoice_Year'] = sales_data['Invoice Date'].dt.year
sales_data['Invoice_Month'] = sales_data['Invoice Date'].dt.month
sales_data['Invoice_Quarter'] = sales_data['Invoice Date'].dt.quarter

# Selecting relevant columns
Yearly_Sales = sales_data[['Custkey', 'DateKey', 'Discount Amount', 'Invoice Date',
                           'Invoice Number', 'Item Class', 'Item', 'Line Number',
                           'List Price', 'Order Number', 'Promised Delivery Date', 'Sales Amount',
                           'Sales Amount Based on List Price', 'Sales Cost Amount',
                           'Sales Margin Amount', 'Sales Price', 'Sales Quantity', 'U/M', 'Invoice_Year']]

# Filter data for years 2017, 2018, and 2019
filtered_data = sales_data[sales_data['Invoice_Year'].isin([2017, 2018, 2019])]

# Group by year and sum Sales Amount
yearly_sales = filtered_data.groupby('Invoice_Year')['Sales Amount'].sum().reset_index()

# Exploratory Data Analysis (EDA)
# Descriptive statistics
print(filtered_data.describe())

```

```python
# Yearly Sales Analysis
plt.figure(figsize=(10, 6))
sns.barplot(x='Invoice_Year', y='Sales Amount', data=yearly_sales, palette='viridis')
plt.title('Yearly Sales Amount (2017-2019)')
plt.xlabel('Year')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
# Group by Invoice Date, Invoice Year, and Invoice Month, and sum the Sales Amount
monthly_sales = sales_data.groupby(['Invoice Date','Invoice_Year','Invoice_Month'])['Sales Amount'].sum().reset_index()

# Plotting the line plot
plt.figure(figsize=(15, 5))
sns.lineplot(x='Invoice_Month', y='Sales Amount', hue='Invoice_Year', data=monthly_sales, palette='bright')
plt.title('Monthly Sales Trend for Each Year')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.legend(title='Year')
plt.show()
```

# **OBSERVATIONS**


*   In 2017, there is Significant fluctuations throughout the year, the Peaks are in March, June, and October, and there is Higher volatility compared to other years.
*   In 2018, there is a Sharp increase in sales from January to March, the Peak
 are in October, highest sales among the three years.
*   In 2019, there is a Lower and more stable sales trend, Slight rise in sales during the middle months. Overall less fluctuation compared to 2017 and 2018.

*  **2018 had the highest sales and was relatively stable compared to the other two years, while 2017 showed high volatility with significant peaks and troughs, and 2019 exhibited the most stable but lower sales trends.**

```python
# Sales Margin Analysis
filtered_data['Sales Margin Percentage'] = (filtered_data['Sales Margin Amount'] / filtered_data['Sales Amount']) * 100
plt.figure(figsize=(10, 6))
sns.histplot(filtered_data['Sales Margin Percentage'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Sales Margin Percentage')
plt.xlabel('Sales Margin Percentage')
plt.ylabel('Frequency')
plt.show()
```

**OBSERVATIONS**

 The histogram shows that while there is a concentration of sales with small positive margins, there is also a significant portion of sales with negative margins, indicating losses. The distribution is heavily skewed, with a long tail on the negative side and a sharp peak on the positive side.
